```python
import pandas as pd
import statsmodels.api as sm
import json
import os
```

```python
df = pd.read_csv("./NLSY.csv")
print(df.shape)
df.head()
```

```python
df=df.dropna(axis=0, how='any')
print(df.shape)
df.head()
```

```python
df_2cols = df.loc[:,["MCS2000","PCS2000"]]
df_2cols.reset_index(inplace=True)
df_2cols.head()
```

```python
df_new=df.drop(columns=["index"])
df_new.head()
```

```python
#reset index adds a index starting from 0.
result = df_new.reset_index().to_json(orient="records")
parsed = json.loads(result)
current_path = os.getcwd()
with open(os.path.join(current_path,"nlsy.json"), 'w') as f:
    json.dump(parsed, f)
print(json.dumps(parsed,indent=2))
```

```python
x=df_new["MCS2000"]
y=df_new["PCS2000"] #match y axis to karen graph. 
x = sm.add_constant(x)
#MCS2000 0.1286 matches MCS2000 .129 from karen
#const   4513.7705 matches 4513.77 from karen
model = sm.OLS(y, x).fit()
predictions = model.predict(x) 
 
print_model = model.summary()
print(print_model)
```
