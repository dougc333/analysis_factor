```python
import pandas as pd
import statsmodels.api as sm
import json
import os
```

```python
df = pd.read_csv("/Users/dc/analysis_factor/datasets_irc_workshop/NLSY.csv")
df.head()
```

```python
df['MCS2000']
```

```python
df=df.dropna(how='all')
x = df['MCS2000']
y = df['PCS2000']
```

```python
df_new = df.loc[:,['MCS2000','PCS2000']]
df_new.head()
```

```python
result = df_new.to_json(orient="records")
parsed = json.loads(result)
current_path = os.getcwd()
with open(os.path.join(current_path,"nlsy.json"), 'w') as f:
    json.dump(parsed, f)
print(json.dumps(parsed,indent=2))
```

```python
x = sm.add_constant(x) # adding a constant
```

```python
model = sm.OLS(y, x).fit()
predictions = model.predict(x) 
 
print_model = model.summary()
print(print_model)
```
